```python
# Attempt to directly scrape Yelp website (initial test)
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# Using selenium webdriver for Safari
driver = webdriver.Safari()

# URL of single Yelp page for initial testing
url = "https://www.yelp.com/biz/my-burma-davis?osq=burma+house#reviews"

driver.get(url)
time.sleep(5)

# Scroll down to trigger lazy loading
for _ in range(3)
driver.find_element(By.TAG_NAME, "body").send_keys(Keys.END)
time.sleep(3) # Sleep to mimic human behavior

# Extract reviews based on given class
review_elements = driver.find_elements(By.XPATH, '//span[contains(@class, "raw__09f24__T4Ezm") and @lang="en"]')
reviews = [review.text for review in review_elements]

driver.quit()

for i, review in enumerate(reviews, 1):
print(f"Review {i}: {review}\n")
```

```python
# Attempt to directly scrape Yelp website
# Using moderately_rated_dollar.csv as an example
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# Load restaurant URLs from CSV
df = pd.read_csv("moderately_rated_dollar.csv")
restaurant_urls = df["yelp_url"].tolist()

driver = webdriver.Safari()

# Create list to store all reviews
all_reviews = []

# Loop through each restaurant URL
for url in restaurant_urls:
    print(f"Scraping reviews for: {url}")

    try:
        driver.get(url)
        time.sleep(5)

        # Scroll down multiple times to load all reviews
        for _ in range(3):
            driver.find_element(By.TAG_NAME, "body").send_keys(Keys.END)
            time.sleep(3)

        # Extract reviews using XPath
        review_elements = driver.find_elements(By.XPATH, '//span[contains(@class, "raw__09f24__T4Ezm") and @lang="en"]')
        reviews = [review.text for review in review_elements]

        # Store reviews with restaurant name
        for review in reviews:
            all_reviews.append({"restaurant_url": url, "review_text": review})

    except Exception as e:
        print(f"Error scraping {url}: {e}")

driver.quit()

reviews_df = pd.DataFrame(all_reviews)
reviews_df.to_csv("moderately_rated_yelp_reviews.csv", index=False)

print("Reviews saved to moderately_rated_yelp_reviews.csv.")
```
